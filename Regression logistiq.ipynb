```python
# Pour diviser la dataset en donnée d'entrainement et de test
from sklearn.model_selection import train_test_split
```

```python
#X = df[['age','sex','cp', 'trestbps','chol','fbs', 'restecg','thalach','exang','oldpeak','slope','ca','thal']].values
X = df.iloc[:, :-1].values
```

```python
# Affichage de X
X
```

```python
# Extraction et Affichage de y
y = df.iloc[:,-1]
y
```

```python
# Division de la dataset
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, random_state=42)
```

REGRESSION LOGISTIQUE

```python
from sklearn.linear_model import LogisticRegression
```

```python
# 1 - Initialisation du modèle
logisticModel = LogisticRegression()
```

```python
# 2- Entrainement du model
logisticModel.fit(X_train, y_train)
```

```python
y_log = logisticModel.predict(X_test)
```

```python
y_log
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
precision_log = accuracy_score(y_test, y_log)
precision_log
```

```python
matrice_log = confusion_matrix(y_test, y_log)
matrice_log
```

```python
cm = ConfusionMatrixDisplay(confusion_matrix=matrice_log)
cm.plot()
```

NORMALISATION DE LA DATASET

```python
#Importation de la librairie
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

Normalisation sur base du df

```python
# Initialisation de la fonction de normalisation
normalized = MinMaxScaler()
normalized.fit_transform(df)
```

```python
#Transformation du df en df_norm
df_norm = normalized.transform(df)
```

```python
# Convertir le tableau numpy en Datafram
df = pd.DataFrame(df)
```

```python
# Regroupement des colonnes du df
column_list = df.columns
column_list
```

```python
df_normalized = pd.DataFrame(df_norm, columns=column_list)
```

```python
df_normalized
```

```python
df_normalized.describe().T
```

```python
sns.pairplot(df_normalized)
```

```python
#Extraction et affichage de X du df_normalized nommé(X_norm)
X_norm = df_normalized.iloc[:, :-1].values
X_norm 
```

```python
# Extraction et Affichage de y du df_normalized nommé(y_norm)
y_norm = df_normalized.iloc[:,-1]
y_norm
```

```python
# Division de la dataset et entrainement
X_norm_train, X_norm_test, y_norm_train, y_norm_test = train_test_split(X_norm,y_norm, test_size=0.1, random_state=42)
```

Regression logistique de la Dataset normalisé

```python
# 1 - Initialisation du modèle
logisticModel = LogisticRegression()
```

```python
# 2- Entrainement du model
logisticModel.fit(X_norm_train, y_norm_train)
```

```python
y_norm_log = logisticModel.predict(X_norm_test)
```

```python
y_norm_log
```

```python
precision_log = accuracy_score(y_norm_test, y_norm_log)
precision_log
```

```python
matrice_log_norm = confusion_matrix(y_norm_test, y_norm_log)
matrice_log_norm
```

```python
cm = ConfusionMatrixDisplay(confusion_matrix=matrice_log_norm)
cm.plot()
```
